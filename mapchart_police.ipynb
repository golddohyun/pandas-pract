## 지도 차트 작성 2

```python
import folium
import pandas as pd
```

```python
police=pd.read_csv('./data_second/03 지도그리기/2016년_사망교통사고.csv', encoding='cp949')
```

```python
police.head(3)
```

```python
police.columns
```

```python
plc = police[['발생지시도', '발생지시군구','사상자수']]
plc = plc.groupby(['발생지시도']).sum()
plc.head(3)
```

```python
idx =plc['사상자수'].sort_values(ascending=False)[:3]
idx.index
```

```python
m = folium.Map(
  location=[36.5053542,127.7043419],
  zoom_start=7
)

# from folium.plugins import MarkerCluster
# marker_cluster = MarkerCluster().add_to(m)

for city in idx.index :
    lac=police[police['발생지시도'] == city][0:1]['위도']
    loc=police[police['발생지시도'] ==city][0:1]['경도']
    folium.Marker(
    location=[lac, loc],
    popup= f'{city} {idx[city]:,}명',
    #popup= city+' '+format(idx[city], ',')+'명',
    icon=folium.Icon(color='red',icon='thumbs-up'),
  ).add_to(m)
    
    folium.CircleMarker(
    [lac,loc],
    radius=idx[city]/20,
    color='#ffffgg',
    fill_color='#fffggg'
    ).add_to(m)
```
